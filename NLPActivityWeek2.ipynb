```python
# E5: Named Entity Recognition
import nltk
nltk.download('averaged_perceptron_tagger')  #figures out how positive or negative text is

```

```python
# E5: NLTK libraries for NER
from nltk import word_tokenize, pos_tag, ne_chunk   # splits up, label parts of speech and chunks parts of speech
from nltk.tree import Tree    # tells you which parts of the sentence have branches.  In this case, Facility and GPE

input_sentence = 'I study at Edge Hill University, which is located in Ormskirk.'


# perform tokenisation, pos tagging and ner in only one line of code
# The NLTK NER identifies the following named entities: Organisation, Person, 
# Location, Date, Time, Money, Percent, Facility and GPE (geo-political entity)
chunks = ne_chunk(pos_tag(word_tokenize(input_sentence)))
print (chunks)
print(chunks.pretty_print())
```

```python
# iterate through chunks
for ne in chunks:
  # if current chunk is a branch of a tree
  # then print it
  if type(ne) == Tree:
    print('Named Entities:\t', ne)
```
